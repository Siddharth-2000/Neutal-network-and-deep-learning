```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import keras
from keras.models import Sequential
from keras.layers import Dense

# Load the Breast Cancer dataset
data = pd.read_csv('breastcancer.csv')
# Assuming your dataset has columns for features and a column for the target variable named 'diagnosis'
X = data.drop('diagnosis', axis=1).values
Y = data['diagnosis'].map({'M': 1, 'B': 0}).values  # Convert diagnosis to binary labels (1 for malignant, 0 for benign)

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=87)

# Normalize the data using StandardScaler
sc = StandardScaler()
X_train_normalized = sc.fit_transform(X_train)
X_test_normalized = sc.transform(X_test)

# Build the neural network model
np.random.seed(155)
my_first_nn = Sequential()
my_first_nn.add(Dense(20, input_dim=X_train.shape[1], activation='relu'))  # Hidden layer
my_first_nn.add(Dense(1, activation='sigmoid'))  # Output layer
my_first_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
my_first_nn_fitted = my_first_nn.fit(X_train_normalized, Y_train, epochs=100, initial_epoch=0, verbose=0)

# Evaluate and report the accuracy on the test data
loss, accuracy = my_first_nn.evaluate(X_test_normalized, Y_test)
print("Accuracy on Test Data:", accuracy*100)
```
